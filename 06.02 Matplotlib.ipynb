- 히스토그램(histogram)
- 상자수염(box plot)

```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
mpl.rcParams['font.family'] ='Malgun Gothic' # 원하는 글꼴
mpl.rcParams['axes.unicode_minus'] =False # 유니코드에서 음수 부호설정
```

```python
# plot(y축), plot(x축, y축)
plt.title('한글쓰면 깨져요')
plt.plot([1, 4, 9, 16])
plt.show()
```

```python
plt.title('x ticks')
#  plot(x축, y축)
plt.plot([10, 20, 30, 40], [1, 4, 9, 16])
plt.show()
```

```python
plt.title('한글 제목')
plt.plot([10, 20, 30, 40], [1, 4, 9, 16])
plt.xlabel('x축 제목')
plt.ylabel('y축 제목')
plt.show()
```

```python
plt.title('한글 제목')
plt.plot([10, 20, 30, 40], [1, 4, 9, 16], 'rs--') # 선색(color), 마커(marker), 선스타일(line style)
plt.xlabel('x축 제목')
plt.ylabel('y축 제목')
plt.show()
```

```python
plt.plot([10, 20, 30, 40], [1, 4, 9, 16], c='b',
        lw=5, ls='--', marker='o', ms=15, mec='g', mew=5, mfc='r')
plt.show()
```

```python
plt.plot([10, 20, 30, 40], [1, 4, 9, 16], c='b',
        lw=5, ls='--', marker='o', ms=15, mec='g', mew=5, mfc='r')
# 축범위
plt.xlim(0, 50)
plt.ylim(-10, 30)
plt.show()
```

```python
x = np.linspace(-np.pi, np.pi, 256)
y = np.cos(x)

plt.plot(x, y)
plt.yticks([-1, 0, 1], ['low', 'zero', 'high'])
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
plt.grid(True)

plt.show()
```

```python
# 여러개의 선 그래프
t = np.arange(0., 5., 0.2)
plt.plot(t, t, 'r--', t, 0.5 * t**2, 'bs:', t, 0.2 * t**3, 'g^-')
plt.show()
```

```python
x = np.linspace(-np.pi, np.pi, 256)
c, s = np.cos(x), np.sin(x)
plt.plot(x, c, ls='--', label='cosine')
plt.plot(x, s, ls=':', label='sine')
plt.legend(loc=0)
plt.show()
```
